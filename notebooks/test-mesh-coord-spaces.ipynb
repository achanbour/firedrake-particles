```python
from firedrake import *
import numpy as np
```

# Mesh

```python
mesh = UnitSquareMesh(2,2)

"""
- SpatialCoordinate(mesh) returns symbolic variables representing the coordinates of the mesh (UFL expression)
- mesh.coordinates encodes the geometric map from the reference cell (Function)
"""

dim = mesh.geometric_dimension
print(dim)

print(SpatialCoordinate(mesh))
x, y = SpatialCoordinate(mesh)
print(x, y)

mesh_coord_func_space = mesh.coordinates.function_space()
print(mesh_coord_func_space)
```

# Mapping from reference space to physical space

```python
# Pick out one cell in the mesh
cell_num = 0
cell_dofs = coords.cell_node_map().values[cell_num, :] # global DoFs associated with the cell
cell_coords = coords.dat.data_ro[cell_dofs, :] # global coords. of the cell vertices

print(cell_coords)
```

```python
# Define a point in reference coordinates
ref_point = np.array([1/3, 1/3])

# Evaluate the local basis functions at that reference point
# using known basis functions for the CG1 finite element on a triangle
basis_values = np.array([
    1-ref_point[0]-ref_point[1],
    ref_point[0],
    ref_point[1]
])

# Map to global space x = G(X)
global_point = basis_values @ cell_coords 
```
